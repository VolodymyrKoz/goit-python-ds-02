Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
grouped_data = python_data.groupby("Должность")
grouped_data
# for name, group in grouped_data:
#     print(f"Група: {name}")
#     print(group)
#     print("\n")
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
aggregated_data = grouped_data['Зарплата.в.месяц'].agg(['min', 'max'])
aggregated_data
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    return row['Зарплата.в.месяц'].mean()

aggregated_data['avg'] = grouped_data.apply(fill_avg_salary)
aggregated_data
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
avg_statistics = aggregated_data["avg"].describe()
avg_statistics

```

Збережіть отриману таблицю в CSV файл

```python
aggregated_data.to_csv('aggregated_salary.csv')
avg_statistics.to_csv("avg_statistic.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
for position in aggregated_data.index:
    plt.plot(['min', 'max', 'avg'], aggregated_data.loc[position], label=position)

plt.xlabel('Salary Type')
plt.ylabel('Salary')
plt.title('Зарплата для різних посад за типом')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
aggregated_data.plot(kind='barh', y=['min', 'max', 'avg'], stacked=True)
plt.xlabel('Salary')
plt.ylabel('Position')
plt.title('Зарплата для різних посад за типом (Горизонтальна стовпчаста діаграма)')
plt.legend()
plt.show()
```
