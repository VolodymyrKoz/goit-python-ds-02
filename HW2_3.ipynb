```python
author_books_count = df.groupby("author")["name"].agg(['count']).reset_index()

max_books_author_index = author_books_count['count'].idxmax()
author_with_max_books = author_books_count.loc[max_books_author_index]
print(author_with_max_books)
print(author_books_count.shape)
author_books_count
```

Який автор має найбільше книг? Відповідь: Jeff Kinney
Скільки книг цього автора? Відповідь: 12

```python
author_avg_rating = df.groupby('author')['user_rating'].agg(['mean']).reset_index()
author_avg_rating = author_avg_rating.rename(columns={'mean': 'avg_rating'})
index = author_avg_rating['avg_rating'].idxmin()
author_with_min_avg_rating = author_avg_rating.loc[index, 'author']
avg_rating = author_avg_rating.loc[author_avg_rating['author'] == author_with_min_avg_rating, 'avg_rating'].values[0]
print(author_with_min_avg_rating)
print(avg_rating)
author_avg_rating
```

У якого автора середній рейтинг мінімальний? Відповідь: Donna Tartt
Який у цього автора середній рейтинг? Відповідь: 3.9

```python
new_df = pd.concat([author_books_count, author_avg_rating["avg_rating"]], axis=1)
new_df
```

```python
new_sorted_df = new_df.sort_values(by=["count", "avg_rating"])
new_sorted_df
```

Який автор перший у списку? Відповідь: Muriel Barbery

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(new_sorted_df['count'], new_sorted_df['avg_rating'], color='blue', alpha=0.7)
plt.xlabel('Кількість книг')
plt.ylabel('Середній рейтинг')
plt.title('Залежність між кількістю книг та середнім рейтингом для кожного автора')
plt.grid(True)
plt.show()
```
